```python
import pandas as pd
```

```python
df = pd.read_excel('data/dataset-1.xlsx')
#df = pd.read_excel('data/dataset-1.xlsx', parse_dates=['Дата'])
df
```

```python
#df['Дата'] = pd.to_datetime(df['Дата'])     #1970-01-01 00:00:00.020180902
f = lambda d: pd.to_datetime(d, format='%Y%m%d')
df['Дата'] = df['Дата'].apply(f)    #1970-01-01 00:00:00.020180902
df
```

```python
sept_df = df[df['Дата'].dt.month < 10]
sept_df
```

```python
okt_df = df[df['Дата'].dt.month == 10]
okt_df
```

```python
nov_df = df[df['Дата'].dt.month == 11]
nov_df
```

```python
session_sept_unique = sept_df['Session ID'].unique().size
session_okt_unique = okt_df['Session ID'].unique().size
session_nov_unique = nov_df['Session ID'].unique().size
print(session_sept_unique)
print(session_okt_unique)
print(session_nov_unique)
print(df['Session ID'].unique().size)
```

```python
mau_sept = sept_df['Client ID'].unique().size
mau_okt = okt_df['Client ID'].unique().size
mau_nov = nov_df['Client ID'].unique().size
print(mau_sept)
print(mau_okt)
print(mau_nov)
```

```python
sum_revenue_sep = sept_df['Доход'].sum()
sum_revenue_okt = okt_df['Доход'].sum()
sum_revenue_nov = nov_df['Доход'].sum()
print(sum_revenue_sep)
print(sum_revenue_okt)
print(sum_revenue_nov)
```

```python
arpu_sept = sum_revenue_sep / mau_sept
arpu_okt = sum_revenue_okt / mau_okt
arpu_nov = sum_revenue_nov / mau_nov
print(arpu_sept)
print(arpu_okt)
print(arpu_nov)
```

```python
dau = sept_df[sept_df['Дата'].dt.day == 12]['Client ID'].unique().size
dau
```

```python
plat_users_sept = sept_df[sept_df['Доход'] > 0]['Client ID'].unique().size
plat_users_okt = okt_df[okt_df['Доход'] > 0]['Client ID'].unique().size
plat_users_nov = nov_df[nov_df['Доход'] > 0]['Client ID'].unique().size

arppu_sept = sum_revenue_sep / plat_users_sept
arppu_okt = sum_revenue_okt / plat_users_okt
arppu_nov = sum_revenue_nov / plat_users_nov
print(arppu_sept)
print(arppu_okt)
print(arppu_nov)
```

```python
sess_per_user_sept = session_sept_unique / mau_sept
sess_per_user_okt = session_okt_unique / mau_okt
sess_per_user_nov = session_nov_unique / mau_nov
print(sess_per_user_sept)
print(sess_per_user_okt)
print(sess_per_user_nov)
```
