```python
import pandas as pd
```

```python
df = pd.read_excel('data/Домашнее задание.xlsx', parse_dates=['Дата'])
df
```

```python
corr_dau = df['Пользователи'].corr(df['Выручка, RUB.1'])
corr_new = df['Новые пользователи'].corr(df['Выручка, RUB.1'])
corr_returned = df['Вернувшиеся'].corr(df['Выручка, RUB.1'])
corr_pay_users = df['Количество платящих пользователей'].corr(df['Выручка, RUB.1'])
corr_buy = df['Количество покупок'].corr(df['Выручка, RUB.1'])
corr_buy_per_pay_user = df['Количество покупок на платящего пользователя'].corr(df['Выручка, RUB.1'])
corr_sr_revenue_per_user = df['Средняя выручка на пользователя, RUB'].corr(df['Выручка, RUB.1'])
corr_sr_revenue_per_pay_user = df['Средняя выручка на платящего пользователя, RUB'].corr(df['Выручка, RUB.1'])
corr_cr_voronki = df['CR воронки покупок, %'].corr(df['Выручка, RUB.1'])
corr_errors = df['Ошибки'].corr(df['Выручка, RUB.1'])
corr_csi = df['CSI'].corr(df['Выручка, RUB.1'])

print(corr_dau)
print(corr_new)
print(corr_returned)
print(corr_pay_users)
print(corr_buy)
print(corr_buy_per_pay_user)
print(corr_sr_revenue_per_user)
print(corr_sr_revenue_per_pay_user)
print(corr_cr_voronki)
print(corr_errors)
print(corr_csi)
```
