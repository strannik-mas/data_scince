### Задание 2. Проведите исследование полученных функций. 

1. $5x + 1.5$
2. $2e^{0.2x}-1$
3. $3\sqrt x - 15$
4. $2\log(x) + 5$
5. $-5x^2 - 10x + 100$




```python
from sympy import *
from sympy.plotting import plot
from sympy.solvers.inequalities import solve_univariate_inequality
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x = Symbol('x')
f1 = 5*x + 1.5
f2 = 2*exp(0.2*x) - 1
f3 = 3*sqrt(x) - 15
f4 = 2*log(x) + 5
f5 = -5*x**2 - 10*x + 100
plot(f1)
plot(f2)
plot(f3)
plot(f4)
plot(f5)
```

```python
solve(f1)
```

```python
solve(f2)
```

```python
solve(f3)
```

```python
solve(f4)
```

```python
solve(f5)
```
