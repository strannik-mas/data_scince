```python
from sympy import *
from sympy.plotting import plot
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x = Symbol('x')
f = x**2
plot(f)
```

```python
print(f.subs(x, 100))
print(f.subs(x, -100))
```

```python
f2 = 2*x**2 + 10*x - 12
plot(f2)
```

```python
print(f2.subs(x, 100))
print(f2.subs(x, -100))
```

```python
f3 = -2*x**2 + 10*x - 12
plot(f3)
```

```python
f4 = -2*x**2 + 10*x + 100
plot(f4)
```

```python
f5 = sqrt(x)
plot(f5)
```

```python
f6 = ln(x) - 10
plot(f6)
```

```python
f5.subs(x,-1)
```

```python
plot(f5, (x, 0, 100))
```

```python
f7 = exp(x) + 1
plot(f7)
```
