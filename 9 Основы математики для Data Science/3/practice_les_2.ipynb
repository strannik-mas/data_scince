```python
from sympy import *
from sympy.plotting import plot
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x = Symbol('x')
f = -3*x + 5
f
```

```python
print(f.subs(x, 5/3))
print(f.subs(x, 2))
plot(f)
```

```python
f2 = 3*x**2 + 3*x - 6
print(f2.subs(x, 0.5))
print(f2.subs(x, 7))
plot(f2)
```

```python
f3 = 2*x**4 + x**3 - 6*x**2 + 7*x + 1
print(f3.subs(x, 3))
print(f3.subs(x, -5))
plot(f3)
```

```python
f4 = (x**2)**x
print(f4.subs(x, 0))
print(f4.subs(x, 10))
plot(f4)
```
