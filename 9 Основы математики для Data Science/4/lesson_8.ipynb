```python
from sympy import *
from sympy.plotting import plot
init_printing(use_unicode=False, wrap_line=False, no_global=True)
a2, a1, a0 = symbols('a2 a1 a0')
```

```python
eq1_lp = a2 - a1 + a0 + 1
eq2_lp = 4*a2 + 2*a1 + a0 - 8
eq3_lp = 4*a2 - 2*a1 + a0 + 8
```

```python
nonlinsolve([eq1_lp, eq2_lp, eq3_lp], [a2, a1, a0])
```

```python
x = Symbol('x')
f = -x**2 + 4*x + 4
```

```python
f.subs(x,-1)
```

```python
f.subs(x, 2)
```

```python
f.subs(x, -2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_range = np.linspace(-3, 7, num=100)
fu = lambda x: -x**2 + 4*x + 4
plt.scatter([-1,2,-2], [-1,8,-8], c='r')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.plot(x_range, fu(x_range))
```
