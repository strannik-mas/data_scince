```python
def div():
    for i in range(2):
        try:
            x = int(input("enter a number: "))
            y = int(input("enter another number: "))
            print(x, '/', y, '=', x/y)
        except ValueError as e:
            print(f'exc: {type(e)} e.args: {e.args}')
        except ZeroDivisionError as e:
            print(f'exc: {type(e)} e.args: {e.args}')
            
div()
```

### Задание 1.

Есть файл с протоколом регистраций пользователей на сайте (registrations.txt).<br>
Каждая строка содержит информацию о имени, электронной почте и возрасте человека. <br><br>


Надо проверить данные из файла, для каждой строки:
 - присутсвуют все три поля
 - поле имени содержит только буквы
 - поле email содержит @ и .
 - поле возраст является числом от 10 до 99<br>

В результате проверки нужно сформировать два файла
 - registrations_good.log для правильных данных, записывать строки как есть
- registrations_bad.log для ошибочных, записывать строку и вид ошибки.<br>

Для валидации строки данных написать метод, который может выкидывать исключения:
 - НЕ присутсвуют все три поля: ValueError
 - поле имени содержит НЕ только буквы: NotNameError (кастомное исключение)
 - поле email НЕ содержит @ и .(точку): NotEmailError (кастомное исключение)
 - поле возраст НЕ является числом от 10 до 99: ValueError
Вызов метода обернуть в try-except.

```python
### YOUR CODE HERE ###
class NotNameError(Exception):
    pass

class NotEmailError(Exception):
    pass


def parseLine(line):
    name, email, age = line.split(' ')
    if name is None or email is None or age is None:
        raise ValueError('Не все параметры')
    if not name.isalpha():
        raise NotNameError('Имя содержит не только буквы')    
    if '@' not in email or '.' not in email:
        raise NotEmailError(f'Почта не содержит знака @ or . email = {email}')
    if not age.isnumeric or not 10 < int(age) < 99:
        raise ValueError(f'поле возраст НЕ является числом от 10 до 99; age = {age}')
        

norm_lines = []
error_lines = []
with open('registrations_.txt', 'r', encoding='utf8') as ff:
    for line in ff:
        try:
            parseLine(line)
        except ValueError as e:
            if 'возраст' in e.args[0]:
                print(f'Проблемы с возрастом {e} в строке {line}')
            else:
                print(f'ошибка {e} в строке {line}')
            error_lines.append(line)
        except NotNameError as e:
            print(f'Проблемы с именем: {e}, в строке {line}')
            error_lines.append(line)
        except NotEmailError as e:
            print(f'Проблемы с email {e} в строке {line}')
            error_lines.append(line)
        else:
            norm_lines.append(line)
                
with open('registrations_good.log', 'w', encoding='utf-8') as f2:
    f2.writelines(norm_lines)
    
with open('registrations_bad.log', 'w', encoding='utf-8') as f3:
    f3.writelines(error_lines)
                
                
            
```
