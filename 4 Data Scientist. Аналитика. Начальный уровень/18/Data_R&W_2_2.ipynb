# Чтение и запись данных

## Чтение и запись файлов в формате JSON

### Проверка практики

```python
import urllib.request as req
import json
import pandas as pd
```

#### Поиск идентификатора города

```python
url = "https://www.metaweather.com/api/location/search/?query=paris"

with req.urlopen(url) as session:
    response1 = session.read().decode()
    data = json.loads(response1)
data[0]
```

#### Получение истории погоды

```python
url = 'https://www.metaweather.com/api/location/615702/2020/05/28/'

df = pd.read_json(url)
df.head()
```

```python
df['the_temp'].mean()
```

### Запись JSON в файлы

```python
url = 'https://www.metaweather.com/api/location/2122265/2020/06/20/'

with req.urlopen(url) as session:
    response = session.read().decode()
    s = json.loads(response)
s
```

```python
file = open('file.json', 'w')
json.dump(s, file, indent=4)
file.close()
```

```python
with open('file.json', 'w') as file:
    json.dump(s, file, indent=4)
```

### Чтение файла JSON

```python
with open('file.json', 'r') as file:
    s = json.load(file)

s
```
