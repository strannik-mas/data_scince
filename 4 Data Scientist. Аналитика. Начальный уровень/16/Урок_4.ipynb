## Урок 4. Форматирование дат

В библиотеке pandas есть особый тип данных, позволяющих хранить в нём даты. Преимущества использования отдельного типа данных для дат, а не строк в том, что мы сможем легко переводить даты из одного формата в другой, производить всякие вычисления над датами, например вычесть из даты семь дней, чтобы узнать, какое было число семь дней назад. Со строками проводить такие операции проблематично.

### Преобразование строковых признаков в дату

Полный список методов и полей, доступных при работе с датами, вы можете посмотреть в документации pandas: https://pandas.pydata.org/pandas-docs/stable/reference/series.html#datetimelike-properties

```python
df = pd.read_csv('Practice2_1.csv', parse_dates=['Date'], dayfirst=False)
df.head(30)
```

```python
df['Day_of_week'] = df['Date'].dt.day_name()
df.head()
```
